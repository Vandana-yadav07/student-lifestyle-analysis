```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Untitled form.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()

print("Current Columns:", df.columns.tolist())

stress_col = [c for c in df.columns if 'Stress' in c][0]
sleep_col = [c for c in df.columns if 'sleep' in c.lower()][0]

df[sleep_col] = pd.to_numeric(df[sleep_col], errors='coerce')
df[stress_col] = pd.to_numeric(df[stress_col], errors='coerce')

print(f"Mean Sleep: {df[sleep_col].mean():.2f}")
print(f"Mean Stress: {df[stress_col].mean():.2f}")
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x='Average hours of sleep per night',y='Current CGPA/Percentage',data=df)
plt.title("Sleep VS Academic Performance")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between All Factors")
plt.show()
```

```python

stress_col = [c for c in df.columns if 'Stress' in c][0]
cgpa_col = [c for c in df.columns if 'CGPA' in c][0]

df[stress_col] = pd.to_numeric(df[stress_col], errors='coerce')
df[cgpa_col] = pd.to_numeric(df[cgpa_col], errors='coerce')

clean_df = df.dropna(subset=[stress_col, cgpa_col])

plt.figure(figsize=(12, 6))
sns.boxplot(x=stress_col, y=cgpa_col, data=clean_df, palette="Set2")
plt.title("CGPA Distribution by Stress Level")
plt.show()
```
